```python
from tabula.io import read_pdf
import pandas as pd
import numpy as np
import re
import csv
import time
import requests

a = pd.read_csv("../input/2019-barcelona-data/2019_divter.csv")
c = a["URL_FITXA_DIVISIO_TERRITORIAL"].to_numpy()
b = []
n = True
nn = 0
```

```python
for i in c:
    successful = False
    print(nn)
   
    try:
        df = read_pdf(i, pages='all')[0]
        df = df.iloc[:, [1,]]
        if n == True and nn < 47 :
            df.drop([0, 1], axis=0, inplace=True)
            n = False
        if nn == 56 or nn == 68 or nn == 72:
            df.drop([0, 1], axis=0, inplace=True)
            n = False
        print(df)
        ind = (int(str(df.iloc[0].values[0]).replace(".", "").split(" ")[0]) - int(str(df.iloc[33].values[0]).replace(".", "").split(" ")[0])) / int(str(df.iloc[0].values[0]).replace(".", "").split(" ")[0]) * 100
        ind = round(ind, 2) * float(str(df.iloc[34].values[0]).replace(",", ".").split(" ")[0])
        ind = round(ind, 2) * float(str(df.iloc[26].values[0]).replace(",", ".").split(" ")[0])
        ind = (round(ind, 2) / 320161) * 100
        sol = float(df.iloc[28].values[0].replace(",", ".").split(" ")[0])
        pob = float(df.iloc[10].values[0].replace(",", ".").split(" ")[0])
        b.append([round(ind, 2), sol, pob])
        nn += 1
        successful = True
        if  16 <= nn < 18:
            n = True
        if nn == 47:
            n = False
    except IndexError:
        nn += 1
        n = True
        pass

        
            
```
